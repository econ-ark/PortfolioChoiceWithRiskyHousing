```python
def create_dict(T_cycle=20, T_retire=10, PermShkStd=0.1029563, PermShkCount=7, RiskyAvg=1.08 , RiskyStd=0.157, RiskyCount=7, HouseAvg=1.0, HouseStd=0.2, HouseShkCount=7):

    dictionary = init_portfolio_risky_housing.copy()
    dictionary["T_retire"] = T_retire
    dictionary["T_cycle"] = T_cycle
    dictionary["T_age"] = T_cycle + 1
    dictionary["LivPrb"] = [1.0] * T_cycle
    dictionary["PermGroFac"] = [1.0] * T_cycle
    dictionary["PermShkStd"] = [PermShkStd] * T_cycle
    dictionary["PermShkCount"] = PermShkCount
    dictionary["TranShkStd"] = [0.0] * T_cycle
    dictionary["TranShkCount"] = 1
    dictionary["HseGroFac"] = 1.01 * T_cycle
    dictionary["RiskyAvg"] = RiskyAvg
    dictionary["RiskyStd"] = RiskyStd
    dictionary["RiskyCount"] = RiskyCount
    dictionary["HouseAvg"] = HouseAvg
    dictionary["HouseStd"] = HouseStd
    dictionary["HouseShkCount"] = HouseShkCount
    dictionary["RentPrb"] = [0.0] * T_retire + list(np.linspace(0,1,T_cycle-T_retire))

    return dictionary
```
